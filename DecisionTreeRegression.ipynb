```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('/home/gxuvimr/Documents/Machine Learning A-Z Template Folder/Part 2 - Regression/Section 8 - Decision Tree Regression/Position_Salaries.csv')

X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

```python
from sklearn.tree import DecisionTreeRegressor
clf = DecisionTreeRegressor()
clf.fit(X, y)
```

```python
y_pred = clf.predict(np.array([6.5]).reshape(1,-1))
print(y_pred)
```

```python
# Visualizing

plt.scatter(X, y, color='red')
plt.plot(X, clf.predict(X), color='blue')
plt.title('Decision Tree Regression')
plt.xlabel('Employee Position')
plt.ylabel('Salary')
plt.show()

# This is not correct as we are visualizing a non-continuos(discrete) regression model in low resolution (resolution by default).
# That's why we were getting a continuous curve there and not here.
```

```python
# Visualization in a higher resolution

X_grid = np.arange(min(X), max(X), 0.01)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X, y, color='red')
plt.plot(X_grid, clf.predict(X_grid), color='blue')
plt.title('Decision Tree Regression')
plt.xlabel('Employee Position')
plt.ylabel('Salary')
plt.show()

# This is correct, clearly divided into appropriate intervals (using averages).
```
