```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('/home/gxuvimr//Documents/Machine Learning A-Z Template Folder/Part 2 - Regression/Section 6 - Polynomial Regression/Position_Salaries.csv')
dataset.head()
```

```python
# Visualizing Data
plt.scatter(x=dataset.loc[:,'Level'], y=dataset.loc[:, 'Salary'])
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.title('Salary v/s Position')
plt.show()
```

```python
# Note: Here we are writing 1:2 not 1, as we always want our X(features) to be a matrix and not a vector (1-D Matrix)
X = dataset.iloc[:, 1:2]
y = dataset.iloc[:, -1]

# This is done generally, but we need maximum data to train our model, for max accuracy and as levels are different
# '''
# from sklearn.model_selection import train_test_split
# X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
# '''
```

```python
from sklearn.linear_model import LinearRegression
clf = LinearRegression()
clf.fit(X, y)

# Now will transform this Linear model into a Polynomial model
from sklearn.preprocessing import PolynomialFeatures
# PolynomialFeatures would transform our matrix of features (X), into a polynomial matrix of features
# Assuming quadratic relationship
poly_reg = PolynomialFeatures(degree=2)
X_poly = poly_reg.fit_transform(X)

clf1 = LinearRegression()
clf1.fit(X_poly, y)
y_pred = clf1.predict(X_poly)
```

```python
# Let's visualize our assumptions
plt.scatter(X, y, color='red')
plt.plot(X, y_pred, color='blue')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.title('Position v/s Salary')
plt.show()
```

```python
# Assuming a cubic relationship
poly_reg = PolynomialFeatures(degree=3)
X_poly = poly_reg.fit_transform(X)

clf1 = LinearRegression()
clf1.fit(X_poly, y)
y_pred = clf1.predict(X_poly)

plt.scatter(X, y, color='red')
plt.plot(X, y_pred, color='blue')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.title('Position v/s Salary')
plt.show()
```
